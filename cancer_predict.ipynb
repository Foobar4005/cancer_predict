# __Dataset__


## Explaining the dataset
- Using the Wisconsin Breast Cancer dataset

- Dataset retrieved from this link: https://www.kaggle.com/datasets/uciml/breast-cancer-wisconsin-data?resource=download

- Features are computed from a digitized image of a fine needle aspirate (FNA) of a breast mass.  They describe characteristics of the cell nuclei present in the image.


<br>The dataset contains ten features for each cell nucleus:
- radius (mean of distances from center to points on the perimeter) 
- texture (standard deviation of gray-scale values) 
- perimeter 
- area 
- smoothness (local variation in radius lengths) 
- compactness (perimeter^2 / area - 1.0) 
- concavity (severity of concave portions of the contour) 
- concave points (number of concave portions of the contour) 
- symmetry 
- fractal dimension ("coastline approximation" - 1)



<br>For each feature, there are three statistics:
- mean
- se (standard error)
- worst (largest)

<br>The remaining columns are the id and the diagnosis
- Diagnosis is either '1' (malignant) or '0' (benign)



```python
#Loading in dataset
df = pd.read_csv('cancer_data.csv')

#Cleaning
df.drop('Unnamed: 32', axis=1, inplace=True)#Removes useless column

#Switching malignant and benign symbols with 0 and 1; not necessary but for personal preference
df['diagnosis'] = df['diagnosis'].replace('M', 1)
df['diagnosis'] = df['diagnosis'].replace('B', 0)

pd.set_option('display.max_columns', None)#To see all columns

df.head()
```

## Preparing the dataset

### Separating features and labels

__Features (listed previously):__
- radius (mean of distances from center to points on the perimeter) 
- texture (standard deviation of gray-scale values) 
- perimeter 
- area 
- smoothness (local variation in radius lengths) 
- compactness (perimeter^2 / area - 1.0) 
- concavity (severity of concave portions of the contour) 
- concave points (number of concave portions of the contour) 
- symmetry 
- fractal dimension ("coastline approximation" - 1)



<br>__Label: diagnosis ('0' for malignant, '1' for benign)__

```python
#Storing features and labels into separates dataframes
X = df.drop(['id', 'diagnosis'], axis=1)#Features
y = df['diagnosis']#Labels

print('LABELS:')
print(y.head())

print('\nFEATURES:')
X.head()
```

### Splitting dataset into training and testing datasets
Since we don't have a separate test dataset, we split our current dataset into two datasets: training (75% of the original dataset) and testing (25% of the original dataset)

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
#Where:
#X_train are the training features
#X_test are the test features
#y_train are the training labels
#y_test are the test labels
X_train.head()
```

### Min-Max Scaling
For some algorithms, it's important that all feature data is on the same scale or it assigns weights to each feature incorrectly. One method of attaining similar scales is to apply min-max scaling. This makes all feature data on a scale of 0-1 relative to their max and min values.

```python
#Applying min-max scaling
scaler = MinMaxScaler()#Initializing scaler
scaler.fit(X_train)#Fitting scaler
scaler.transform(X_train)#Applying scale to training features
scaler.transform(X_test)#Applying scale to test features
```
